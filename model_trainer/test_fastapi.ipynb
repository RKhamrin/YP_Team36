```python
import pandas as pd 
import numpy as np 
import requests

import json
from io import StringIO
```

## Show_example

```python
resp = requests.get('http://127.0.0.1:8000/api/models/show_example')

content = StringIO(resp.content.decode("utf-8"))
data = pd.read_csv(content).reset_index().drop(columns=['index'])
data.head()
```

## Fit

```python
url = f'http://127.0.0.1:8000/api/models/fit?jsonfile={json.dumps({"model_id": "logistic_123", "hyperparameters": {}})}'

files = {
        'data': ('data/data_sample.csv', open('data/data_sample.csv', 'rb'), 'multipart/form-data')
}

resp = requests.post(url=url, files=files)
print(resp.json())
```

## Set_model

```python
url = f'http://127.0.0.1:8000/api/models/set_model'
resp = requests.post(url=url, json={"id": "logistic_123"})
print(resp.json())
```

## Show_models

```python
url = f'http://127.0.0.1:8000/api/models/show_models'
resp = requests.get(url=url)
print(resp.json())
```

## Predict

```python
url = f'http://127.0.0.1:8000/api/models/predict?jsonfile={json.dumps({"model_id": "logistic_123"})}'

files = {
        'data': ('data/data_sample.csv', open('data_sample_predict.csv', 'rb'), 'multipart/form-data')
}

resp = requests.post(url=url, files=files)
content = StringIO(resp.content.decode("utf-8"))
data = pd.read_csv(content).reset_index().drop(columns=['index'])
data.head()
```
